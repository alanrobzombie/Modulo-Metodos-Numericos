```python
import numpy as np
import matplotlib.pyplot as plt

def eval(x):
    return 3*x**2+2*x-3
    
def bisec(a, b, tol=1e-6):
    sol=0
    for _ in range(40000):
        x_r=(a+b)/2
        if(eval(a)*eval(x_r)<0):
            b=x_r
        elif(eval(a)*eval(x_r)>0):
            a=x_r
        else:
            return x_r
        sol=x_r
    return sol
    
print(bisec(0, 1))

x = np.linspace(0,1.5, 10000)
y = 3*x**2+2*x-3

lienzo = plt.figure(figsize=(14,7), dpi=200)
ejes = lienzo.add_axes([0.1, 0.1, 0.8, 0.8])
ejes.plot(x, y, 'g')
ejes.set_xlabel('x', fontsize=19, c='blue')
ejes.set_ylabel('y', fontsize=19, c='red')
ejes.set_xlim(0.7, 0.8)
ejes.set_ylim(0, 1)
ejes.grid(True, ls='--')


```
